# FE 3) Ordinal Number Encoding or Label Encoding

```python
import pandas as pd
import datetime
```

```python
# create a variable with dates, and from that extract the weekday
df_base = datetime.datetime.today()

#create a list of dates with 10 days difference from today
df_date_list = [df_base - datetime.timedelta(days=x) for x in range(0,10)]

#then transform the list into a dataframe
df = pd.DataFrame(df_date_list)
df.columns = ['day']
df
```

```python
df['day_of_week'] = df['day'].dt.day_name() #generated day name for each date
df.head(10)
```

```python
weekday_map = {'Monday':1,
               'Tuesday':2,
               'Wednesday':3,
               'Thursday':4,
               'Friday':5,
               'Saturday':6,
               'Sunday':7
    
}

df['day_ordinal'] = df.day_of_week.map(weekday_map)
df.head(10) #assigned rank to each day
```
